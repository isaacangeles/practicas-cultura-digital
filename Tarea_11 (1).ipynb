## Parte 3: Condicionales en Python

### Ejercicio 1
Pide la hora (0-23). Si es menor que 12 muestra 'Es mañana', si no 'Es tarde o noche'.


```python
pozole=int(input("dime la hora del dia"))
if(pozole<13):
  print("es de mañana")
elif(pozole>12):
  print("es de tarde/noche")



```

### Ejercicio 2
Pide dos números y muestra cuál es mayor o si son iguales.


```python

tilin=int(input("dime un numero"))
sonic=int(input("dime otro numero"))
if(tilin<sonic):
  print(tilin, "es mayor que", sonic)
elif(sonic<tilin):
  print(sonic, "es mayor que", tilin)
else:
  print("son iguales")

```
